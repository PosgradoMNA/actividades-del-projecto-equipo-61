**1 Caso del negocio.** 
Entender claramente el alcance, requerimientos y objetivos del proyecto.


**2 Entender los datos.** 
Inicia desde el proceso de generar los datos y cubre las fases exploratorias iniciales para tratar de detectar patrones y tendendencia en la información recabada

**3 Preparación de los datos.** 
En esta fase se limpian los datos,se toman las decisiones sobre los datos faltantes, se normalizan y en general el producto final es el conjunto de datos final a utilizar en el proyecto

**4 Generar modelos.** 
En esta fase comienza un proceso que conlleva multiples iteraciones de manipulaciones de los datos y creación de modelos. Es muy extraño que en un proyecto un analista se quede con un único modelo.

**5 Evaluar modelos.** 
Una vez que se tienen varios modelos tecnicamente correctos, en esta fase se comparan utilizando criterios rendimiento previamente elegidos (el criterio va a ir ligado al tipo de problema). De la comparación si se llega a un modelo que se considera adecuado se pasa a la siguiente fase y si no se puede ir hacia atrás e intentar nuevos callejones.

**6 Desplegar.** 
En esta fase se utilza el modelo definido para la resolución del caso del negocio. De igual manera, se puede aprovechar cualquier información descubirta en el proceso para mejorar el negocio. Adicionalmente, en esta fase se implementan los procesos de control del modelo en la producción final para evitar fallos.
